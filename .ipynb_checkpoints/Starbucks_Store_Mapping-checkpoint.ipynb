## 스타벅스 매장 찾기

http://www.istarbucks.co.kr

- STORE > 매장 찾기
- 지역검색 > 서울 > 전체

````python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
from pprint import pprint
````

````python
data = {
    'ins_lat' : '37.56682', # 지정한 위도와 경도에서 가까운 순으로 나열
    'ins_lng' : '126.97865',
    'p_sido_cd' : '01', # 01=서울시, 08=경기 ... 16=제주
    'p_gugun_cd' : '', # 세부지역 (지정하지 않으면 시/도 전체)
    'in_biz_cd':'',
    'set_date':'',
    'iend':'1000',
}

SB_URL = 'https://www.istarbucks.co.kr/store/getStore.do'
r = requests.post(SB_URL, data=data)

r.text[:1000]
````

## JSON to DataFramem

````python
jo = json.loads(r.text) # str to json
````

````python
df = json_normalize(jo, 'list') # json to DataFrame
````

````python
len(df) # 서울 총 457개 매장
````

````python
df.columns # 119개 컬럼
````

````python
# 주요한 컬럼 몇 가지 선택
               
df = df[['s_name', 'lat', 'lot', 'sido_name', 'gugun_name', 'doro_address', 'tel']]
df.head(20)
````

## 위도 경도 데이터 타입 변환

````python
df.dtypes
````

#### lat, lot -> float으로 전환

````python
df['lat'] = df['lat'].astype(float)
df['lot'] = df['lot'].astype(float)
````

````python
df.dtypes
````

## folium

https://github.com/python-visualization/folium

파이썬 지리정보 시각화 모듈 (문서: https://folium.readthedocs.io )

````python
import folium

# 서울 시청
coord_city_hall = (37.56629, 126.979808)
map_osm = folium.Map(location=(coord_city_hall))
map_osm
````

````python
# zoop start
map_osm = folium.Map(location=(coord_city_hall), zoom_start=17)
map_osm
````

````python
# 지도 title 지정
map_osm = folium.Map(location=(coord_city_hall), zoom_start=17, tiles='Stamen Toner')
map_osm
````

````python
map_osm = folium.Map(location=(coord_city_hall),
                     zoom_start=17,)
folium.Marker(coord_city_hall, popup='시청').add_to(map_osm)
map_osm
````

````python
map_osm = folium.Map(location=(coord_city_hall),
                     zoom_start=12,)
for ix, row in df.iterrows():
    coord_loc = (row['lat'], row['lot'])
    folium.Marker(coord_loc, popup=row['s_name']).add_to(map_osm)
    
map_osm
````

## 제주도 스타벅스 매장 지도

````python
coord_halla = (33.361666, 126.529166)

data = {
    'ins_lat': '33.4996213', # 제주 시청의 좌표
    'ins_lng': '126.5311884',
    'p_sido_cd':'16', # 01=서울시, 08=경기 ... 16=제주
    'p_gugun_cd':'',  # 세부지역 (지정하지 않으면 시/도 전체)
    'in_biz_cd':'',
    'set_date':'',
    'iend':'1000',
}
    
url = 'https://www.istarbucks.co.kr/store/getStore.do'
r = requests.post(url, data=data)
df = json_normalize(json.loads(r.text), 'list')

map_osm = folium.Map(location=coord_halla, zoom_start=10)
for ix, row in df.iterrows():
    location = (row['lat'], row['lot'])
    folium.Marker(location, popup=row['s_name']).add_to(map_osm)

map_osm
````
